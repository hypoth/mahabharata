```python
import re
from pathlib import Path
import pandas as pd
import wikipedia
```

```python

## Reading metadata
directory = Path("../text/KMGanguli")
directory_wikipedia = Path("../text/Wikipedia")
csv_metadata = directory/"metadata_books.csv"

df = pd.read_csv(csv_metadata, header=0, sep=";", names=['book_number', 'title', 'sp', 'contents', 'comments'])

## Split title into separate fields
title_list = []
for title in df['title']:
    title_list = title_list + [re.split(r"\[|\]|\(|\)", title)]

dt = pd.DataFrame(title_list, columns=['book_number', 'book', '3', 'wiki_link', '5', 'description', '7'])

## Split sub parvas into start and end 
sp_list = []
for sp in df['sp']:
    sp_list = sp_list + [re.split(r"–|-", sp)]

sp = pd.DataFrame(sp_list, columns=['start_chapter', 'end_chapter'])
dt.drop(['3', '5', '7', 'book_number'], axis=1, inplace=True)
df.drop(['title', 'comments', 'sp'], axis=1, inplace=True)

## Combine the dataframes
mtdt = pd.concat([dt, sp, df], axis=1)
```

```python
def wiki_summary(key):
    # print(key)
    summary =  wikipedia.summary(key)
    return summary

mtdt['summary'] = mtdt['book'].apply(lambda x: wiki_summary(x))
# mtdt.summary
```

```python
mtdt.to_csv(directory_wikipedia/"wikipedia_parva_summary.csv", sep=";")
for i, row in mtdt.iterrows():
    print(row['book'], "\n", row['summary'], "\n\n\n")
```
